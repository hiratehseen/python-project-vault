# Analyzing Apple Stock Prices with AutoTS

In this code, we'll walk through an example of using the AutoTS library to analyze and forecast Apple stock prices. AutoTS is an automated time series forecasting library that simplifies the process of building and evaluating forecasting models. First, we need to install the required packages.

Next, we import the necessary libraries: numpy for numerical computations, pandas for data manipulation, and plotly.graph_objects for visualizations. Additionally, we load the stock price data for Apple from a CSV file using 'pd.read_csv()' and inspect the first few rows.

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
data = pd.read_csv("/content/drive/MyDrive/Projects/AAPL.csv")
data.head()
```

## Visualizing the Stock Price Data

To gain insights into the stock price data, we create a candlestick chart using the go.Candlestick class from the plotly.graph_objects module. This chart represents the opening, closing, highest, and lowest prices of Apple stock over time. We pass the relevant columns from the data DataFrame to the go.Candlestick constructor and store the resulting figure in the figure variable. Finally, we customize the layout and display the figure.

```python
figure = go.Figure(data=[go.Candlestick(x=data["Date"],
                                        open=data["Open"], high=data["High"],
                                        low=data["Low"], close=data["Close"])])
figure.update_layout(title = "Apple Stock Price Analysis", xaxis_rangeslider_visible=False)
figure.show()
```

This visualization provides an overview of the historical Apple stock prices, allowing us to identify trends, patterns, and price movements.

## Calculating Correlations

To examine the relationships between different columns in the data DataFrame, we calculate the correlation coefficients using the data.corr() method. This function computes the pairwise correlation of columns, excluding missing values, and returns a new DataFrame containing the correlation matrix.

```python
data.corr()
```

The resulting correlation matrix provides insights into the strength and direction of the relationships between variables, such as the correlation between the closing price and other features.

# Forecasting Future Stock Prices

Now, let's use the AutoTS library to forecast future stock prices. We instantiate an AutoTS object, specifying the forecast length (the number of future periods to predict) and the frequency of the data. The ensemble='simple' parameter indicates that we want to use a simple ensemble model.

```python
from autots import AutoTS

# Create an empty DataFrame with the dates for the forecasted period
forecast_dates = pd.date_range(start=data["Date"].iloc[-1], periods=30, freq='B')
forecast_df = pd.DataFrame(index=forecast_dates)

# Initialize and fit the AutoTS model
model = AutoTS(forecast_length=30, frequency='infer', ensemble='simple')
model = model.fit(data, date_col='Date', value_col='Close', id_col=None)

# Generate the forecast
prediction = model.predict()
forecast_df["Forecasted Prices"] = prediction.forecast

print(forecast_df["Forecasted Prices"])
```

The fit() method fits the model to the data, using the specified date column (date_col), value column (value_col), and optional identifier column (id_col). The predict() method generates forecasts based on the trained model, and we store the forecasted values in the forecast variable.
Finally, we print the forecasted values to see the predicted stock prices for the next forecast_length periods. Showing the future stock prices based on the trained model.

# Prediction and Actual Visulization

The visualization uses Plotly's go.Scatter to plot two traces: one for the actual prices and one for the forecasted prices. The actual prices are shown as scattered markers, with the x-axis representing dates and the y-axis representing the historical stock prices. The forecasted prices are displayed as a line plot, connecting the predicted points. The x-values for the forecasted prices come from the future dates, and the y-values come from the predicted stock prices. By comparing the actual and forecasted prices on the same chart, we can visually assess the accuracy of the forecasting model and identify any trends or differences.

```python
# Plot the actual and forecasted prices
figure = go.Figure(layout_yaxis_range=[100, 170])

# Add the actual prices as a scatter plot
figure.add_trace(go.Scatter(x=data["Date"], y=data["Close"], mode="markers", name="Actual"), )

# Add the forecasted prices as a line plot
figure.add_trace(go.Scatter(x=forecast_df.index, y=forecast_df["Forecasted Prices"], mode="lines", name="Forecast"))

# Customize the layout
figure.update_layout(title="Apple Stock Price Forecast",
                    xaxis_title="Date",
                    yaxis_title="Stock Price")

# Show the figure
figure.show()
```

The resulting chart provides an intuitive visual representation of the forecasted stock prices over time. It allows we to see how the predicted prices align with the actual prices and assess the accuracy and performance of the forecasting model.
