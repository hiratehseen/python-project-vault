```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
r = requests.get('https://www.psx.com.pk/market-summary/')
htmlContent = r.content
print(htmlContent)
```

```python
soup = BeautifulSoup(htmlContent)
class_of_table = soup.find('div', {'class' : 'inner-content-box'})
class_of_table
```

```python
print(soup.prettify())
```

```python
first_col = []
current_val = []
for i in class_of_table('div', {'class' : "table-responsive"}):
    table = i.table
#     print(table)
    tableRows = [[td.text for td in row.find_all("td")] for row in table.find_all("tr")[1:]]
#     print(tableRows[1])
    first_col.append(tableRows[1][0])
    current_val.append(tableRows[1][5])
# print(first_col)
# print(current_val)
```

```python
# build df from tableRows and headers
df = pd.DataFrame(current_val[1:], first_col[1:])
print(df)
```
